```python
import pandas as pd
import numpy as np
import glob
import os

mist_dir = r"C:\Users\Lenovo\Downloads\BinaryWorlds\MIST_v1.2_vvcrit0.0_basic_isos"

files = glob.glob(
    os.path.join(mist_dir, "*.iso")
)

print("Files found:", len(files))
```

```python
colnames = [
    "EEP",
    "logAge",
    "initial_mass",
    "star_mass",
    "star_mdot",
    "he_core_mass",
    "c_core_mass",
    "log_L",
    "log_LH",
    "log_LHe",
    "log_Teff",
    "log_R",
    "log_g",
    "surface_h1",
    "surface_he3",
    "surface_he4",
    "surface_c12",
    "surface_o16",
    "log_center_T",
    "log_center_rho",
    "center_gamma",
    "center_h1",
    "center_he4",
    "center_c12",
    "phase"
]
```

```python
all_mist = []

for file in files:

    print("Loading:", os.path.basename(file))

    rows = []

    with open(file) as f:

        for line in f:

            line = line.strip()

            if not line:
                continue

            if line.startswith("#"):
                continue

            parts = line.split()

            if len(parts) != 25:
                continue

            try:
                float(parts[0])
            except:
                continue

            rows.append(parts)

    df = pd.DataFrame(rows, columns=colnames)

    df = df.astype(float)

    # Extract metallicity from filename
    fname = os.path.basename(file)

    if "_feh_p" in fname:
        feh = float(
            fname.split("_feh_p")[1]
                 .split("_")[0]
        )

    elif "_feh_m" in fname:
        feh = -float(
            fname.split("_feh_m")[1]
                 .split("_")[0]
        )

    df["FeH"] = feh

    all_mist.append(df)

mist = pd.concat(
    all_mist,
    ignore_index=True
)

print(mist.shape)
```
