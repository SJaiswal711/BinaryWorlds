```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import norm
from scipy.integrate import trapezoid

# ==========================================
# 1. LOAD CLEAN CSV
# ==========================================

df = pd.read_csv("binary_radius_gap_clean_input.csv")
print("Loaded clean sample:", len(df))

# Keep only the radius-gap regime
df = df[
    (df["Rppri"] > 0.5) & (df["Rppri"] < 6.0) &
    (df["Rpsec"] > 0.5) & (df["Rpsec"] < 6.0)
].copy()

print("Radius-gap sample:", len(df))

# ==========================================
# 2. BASIC CLEANUP
# ==========================================

df = df[(df["p_pri"] + df["p_sec"]) > 0].copy()
df["p_pri"] = df["p_pri"] / (df["p_pri"] + df["p_sec"])
df["p_sec"] = 1.0 - df["p_pri"]

df["delta_radius"] = df["Rpsec"] - df["Rppri"]
df["radius_shift"] = df["Rpsec"] / df["Rppri"]

print("\nHost ambiguity:")
ambiguous = df[(df["p_pri"] > 0.2) & (df["p_pri"] < 0.8)]
print(f"Ambiguous systems: {len(ambiguous)} / {len(df)} = {len(ambiguous)/len(df):.3f}")

print("\nRadius shift summary:")
print(df["radius_shift"].describe())

# ==========================================
# 3. PLOTS OF BASIC DIAGNOSTICS
# ==========================================

outdir = "binary_radius_gap_outputs"
os.makedirs(outdir, exist_ok=True)

plt.figure(figsize=(8,5))
plt.hist(df["p_pri"], bins=20, density=True)
plt.xlabel(r"$p_{\rm pri}$")
plt.ylabel("Density")
plt.title("Host-Probability Distribution")
plt.tight_layout()
plt.savefig(os.path.join(outdir, "p_pri_distribution_clean.png"), dpi=200)
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df["delta_radius"], bins=25, density=True)
plt.xlabel(r"$R_{p,sec}-R_{p,pri}$ [$R_\oplus$]")
plt.ylabel("Density")
plt.title("Radius Inflation")
plt.tight_layout()
plt.savefig(os.path.join(outdir, "delta_radius_distribution_clean.png"), dpi=200)
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(df["Rppri"], df["Rpsec"], c=df["p_pri"], cmap="viridis", alpha=0.85)
plt.plot([0.5, 6], [0.5, 6], "--")
plt.axvspan(1.5, 2.0, alpha=0.12)
plt.axhspan(1.5, 2.0, alpha=0.12)
plt.colorbar(label=r"$p_{\rm pri}$")
plt.xlabel(r"$R_{p,pri}$ [$R_\oplus$]")
plt.ylabel(r"$R_{p,sec}$ [$R_\oplus$]")
plt.title("Primary vs Secondary Host Radii")
plt.tight_layout()
plt.savefig(os.path.join(outdir, "Rppri_vs_Rpsec_clean.png"), dpi=200)
plt.show()

# ==========================================
# 4. BUILD POPULATION DISTRIBUTIONS
# ==========================================

R_grid = np.linspace(0.5, 6.0, 2000)

def safe_sigma(row, mean_col, plus_col, minus_col, floor=0.02):
    vals = []
    for c in [plus_col, minus_col]:
        v = row.get(c, np.nan)
        if pd.notna(v):
            vals.append(abs(v))
    if len(vals) == 0:
        mu = row[mean_col]
        return max(0.10 * abs(mu), floor)
    return max(0.5 * np.mean(vals), floor)

def normalize_pdf(y, x):
    area = trapezoid(y, x)
    if area <= 0 or not np.isfinite(area):
        return np.zeros_like(y)
    return y / area

P_primary_only = np.zeros_like(R_grid)
P_secondary_only = np.zeros_like(R_grid)
P_probabilistic = np.zeros_like(R_grid)

sigma_pri_arr = []
sigma_sec_arr = []

for _, row in df.iterrows():
    Rp_pri = row["Rppri"]
    Rp_sec = row["Rpsec"]

    sigma_pri = safe_sigma(row, "Rppri", "E_Rppri", "e_Rppri")
    sigma_sec = safe_sigma(row, "Rpsec", "E_Rpsec", "e_Rpsec")

    sigma_pri_arr.append(sigma_pri)
    sigma_sec_arr.append(sigma_sec)

    p_pri = row["p_pri"]
    p_sec = row["p_sec"]

    pdf_pri = norm.pdf(R_grid, loc=Rp_pri, scale=sigma_pri)
    pdf_sec = norm.pdf(R_grid, loc=Rp_sec, scale=sigma_sec)

    pdf_pri = normalize_pdf(pdf_pri, R_grid)
    pdf_sec = normalize_pdf(pdf_sec, R_grid)

    P_primary_only += pdf_pri
    P_secondary_only += pdf_sec
    P_probabilistic += p_pri * pdf_pri + p_sec * pdf_sec

P_primary_only = normalize_pdf(P_primary_only, R_grid)
P_secondary_only = normalize_pdf(P_secondary_only, R_grid)
P_probabilistic = normalize_pdf(P_probabilistic, R_grid)

sigma_pri_arr = np.array(sigma_pri_arr)
sigma_sec_arr = np.array(sigma_sec_arr)

# ==========================================
# 5. SIMPLE VALLEY METRIC
# ==========================================

PEAK1 = (0.8, 1.4)
VALLEY = (1.5, 2.0)
PEAK2 = (2.0, 3.5)

def valley_metric(R, P):
    m1 = (R >= PEAK1[0]) & (R <= PEAK1[1])
    mv = (R >= VALLEY[0]) & (R <= VALLEY[1])
    m2 = (R >= PEAK2[0]) & (R <= PEAK2[1])

    peak1 = np.nanmax(P[m1])
    valley = np.nanmin(P[mv])
    peak2 = np.nanmax(P[m2])

    avg_peaks = 0.5 * (peak1 + peak2)
    depth_ratio = valley / avg_peaks if avg_peaks > 0 else np.nan

    return {
        "peak1": peak1,
        "valley": valley,
        "peak2": peak2,
        "avg_peaks": avg_peaks,
        "depth_ratio": depth_ratio
    }

stats_primary = valley_metric(R_grid, P_primary_only)
stats_secondary = valley_metric(R_grid, P_secondary_only)
stats_prob = valley_metric(R_grid, P_probabilistic)

print("\nValley metrics:")
print("Primary-only   :", stats_primary)
print("Secondary-only :", stats_secondary)
print("Probabilistic  :", stats_prob)

# ==========================================
# 6. MONTE CARLO HOST-MARGINALIZED POPULATIONS
# ==========================================

rng = np.random.default_rng(42)
N_MC = 3000
bins = np.linspace(0.5, 6.0, 60)
bin_centers = 0.5 * (bins[:-1] + bins[1:])
bin_width = bins[1] - bins[0]

Rppri = df["Rppri"].to_numpy()
Rpsec = df["Rpsec"].to_numpy()
ppri = df["p_pri"].to_numpy()
sigma_pri_arr = np.array(sigma_pri_arr)
sigma_sec_arr = np.array(sigma_sec_arr)

mc_hists = np.zeros((N_MC, len(bin_centers)))

for i in range(N_MC):
    choose_primary = rng.random(len(df)) < ppri
    means = np.where(choose_primary, Rppri, Rpsec)
    sigmas = np.where(choose_primary, sigma_pri_arr, sigma_sec_arr)

    draws = rng.normal(means, sigmas)
    draws = draws[(draws >= 0.5) & (draws <= 6.0)]

    counts, _ = np.histogram(draws, bins=bins)
    if counts.sum() > 0:
        mc_hists[i] = counts / (counts.sum() * bin_width)

mc_median = np.median(mc_hists, axis=0)
mc_lo = np.percentile(mc_hists, 16, axis=0)
mc_hi = np.percentile(mc_hists, 84, axis=0)

mc_median = normalize_pdf(mc_median, bin_centers)
mc_lo = normalize_pdf(mc_lo, bin_centers)
mc_hi = normalize_pdf(mc_hi, bin_centers)

mc_stats = valley_metric(bin_centers, mc_median)
print("\nMC median valley metrics:", mc_stats)

# ==========================================
# 7. SUMMARY TABLES
# ==========================================

summary = pd.DataFrame([
    {"model": "primary_only", **stats_primary},
    {"model": "secondary_only", **stats_secondary},
    {"model": "probabilistic", **stats_prob},
    {"model": "mc_median", **mc_stats},
])

summary.to_csv(os.path.join(outdir, "valley_summary.csv"), index=False)

pop_df = pd.DataFrame({
    "R_grid": R_grid,
    "P_primary_only": P_primary_only,
    "P_secondary_only": P_secondary_only,
    "P_probabilistic": P_probabilistic,
})
pop_df.to_csv(os.path.join(outdir, "population_pdfs.csv"), index=False)

mc_df = pd.DataFrame({
    "R_bin_center": bin_centers,
    "mc_median": mc_median,
    "mc_lo": mc_lo,
    "mc_hi": mc_hi,
})
mc_df.to_csv(os.path.join(outdir, "mc_population_summary.csv"), index=False)

# ==========================================
# 8. FINAL COMPARISON PLOT
# ==========================================

plt.figure(figsize=(10,6))
plt.plot(R_grid, P_primary_only, label="Primary-only", lw=2)
plt.plot(R_grid, P_secondary_only, label="Secondary-only", lw=2, alpha=0.8)
plt.plot(R_grid, P_probabilistic, label="Probabilistic", lw=2.5)

plt.fill_between(bin_centers, mc_lo, mc_hi, alpha=0.2, label="MC 16–84% band")
plt.plot(bin_centers, mc_median, "--", lw=2, label="MC median")

plt.axvspan(1.5, 2.0, alpha=0.12, label="Gap window")

plt.xlabel(r"Planet Radius [$R_\oplus$]")
plt.ylabel("Normalized Population Density")
plt.title("Binary-Aware Radius Distribution")
plt.xlim(0.5, 6.0)
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(outdir, "binary_aware_radius_distribution.png"), dpi=220)
plt.show()

# ==========================================
# 9. PRINT GLOBAL SUMMARY
# ==========================================

print("\n==============================")
print("GLOBAL SUMMARY")
print("==============================")
print("Clean sample size:", len(df))
print("Ambiguous hosts:", len(ambiguous))
print("Ambiguous fraction:", len(ambiguous)/len(df))
print("Mean p_pri:", df["p_pri"].mean())
print("Median p_pri:", df["p_pri"].median())
print("Mean radius shift:", df["radius_shift"].mean())
print("Median radius shift:", df["radius_shift"].median())
print("Primary-only valley depth ratio:", stats_primary["depth_ratio"])
print("Secondary-only valley depth ratio:", stats_secondary["depth_ratio"])
print("Probabilistic valley depth ratio:", stats_prob["depth_ratio"])
print("MC median valley depth ratio:", mc_stats["depth_ratio"])
print(f"\nSaved outputs to: {outdir}")
```

```python
# ==========================================
# BINARY- AWARE RADIUS GAP ANALYSIS
# NEXT STEP: LOAD CLEAN CSV + STATISTICS
# ==========================================

import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import norm
from scipy.integrate import trapezoid

# ------------------------------------------
# CONFIG
# ------------------------------------------

INPUT_FILE = "binary_radius_gap_clean_input.csv"
OUTDIR = "binary_radius_gap_nextstep"
os.makedirs(OUTDIR, exist_ok=True)

R_MIN, R_MAX = 0.5, 6.0
GRID_N = 2000
MC_REALIZATIONS = 2000
BOOTSTRAP_REALIZATIONS = 2000
RNG_SEED = 42

PEAK1 = (0.8, 1.4)
VALLEY = (1.5, 2.0)
PEAK2 = (2.0, 3.5)

rng = np.random.default_rng(RNG_SEED)

# ------------------------------------------
# HELPERS
# ------------------------------------------

def safe_sigma(row, mean_col, plus_col, minus_col, floor=0.02, frac_fallback=0.10):
    vals = []
    for c in [plus_col, minus_col]:
        v = row.get(c, np.nan)
        if pd.notna(v):
            vals.append(abs(v))
    if len(vals) > 0:
        sigma = 0.5 * np.mean(vals)
    else:
        mu = row.get(mean_col, np.nan)
        sigma = frac_fallback * abs(mu) if pd.notna(mu) and mu != 0 else floor
    return max(float(sigma), floor)

def normalize_pdf(y, x):
    area = trapezoid(y, x)
    if not np.isfinite(area) or area <= 0:
        return np.zeros_like(y)
    return y / area

def valley_metric(R, P):
    m1 = (R >= PEAK1[0]) & (R <= PEAK1[1])
    mv = (R >= VALLEY[0]) & (R <= VALLEY[1])
    m2 = (R >= PEAK2[0]) & (R <= PEAK2[1])

    peak1 = np.nanmax(P[m1]) if np.any(m1) else np.nan
    valley = np.nanmin(P[mv]) if np.any(mv) else np.nan
    peak2 = np.nanmax(P[m2]) if np.any(m2) else np.nan

    avg_peaks = 0.5 * (peak1 + peak2) if np.isfinite(peak1) and np.isfinite(peak2) else np.nan
    depth_ratio = valley / avg_peaks if np.isfinite(valley) and np.isfinite(avg_peaks) and avg_peaks > 0 else np.nan

    return {
        "peak1": peak1,
        "valley": valley,
        "peak2": peak2,
        "avg_peaks": avg_peaks,
        "depth_ratio": depth_ratio,
    }

def build_population(df, R_grid):
    P_primary = np.zeros_like(R_grid, dtype=float)
    P_secondary = np.zeros_like(R_grid, dtype=float)
    P_prob = np.zeros_like(R_grid, dtype=float)

    sigma_pri_list = []
    sigma_sec_list = []

    for _, row in df.iterrows():
        Rp_pri = float(row["Rppri"])
        Rp_sec = float(row["Rpsec"])

        sigma_pri = safe_sigma(row, "Rppri", "E_Rppri", "e_Rppri")
        sigma_sec = safe_sigma(row, "Rpsec", "E_Rpsec", "e_Rpsec")

        p_pri = float(row["p_pri"])
        p_sec = float(row["p_sec"])

        pdf_pri = norm.pdf(R_grid, loc=Rp_pri, scale=sigma_pri)
        pdf_sec = norm.pdf(R_grid, loc=Rp_sec, scale=sigma_sec)

        pdf_pri = normalize_pdf(pdf_pri, R_grid)
        pdf_sec = normalize_pdf(pdf_sec, R_grid)

        P_primary += pdf_pri
        P_secondary += pdf_sec
        P_prob += p_pri * pdf_pri + p_sec * pdf_sec

        sigma_pri_list.append(sigma_pri)
        sigma_sec_list.append(sigma_sec)

    P_primary = normalize_pdf(P_primary, R_grid)
    P_secondary = normalize_pdf(P_secondary, R_grid)
    P_prob = normalize_pdf(P_prob, R_grid)

    return P_primary, P_secondary, P_prob, np.array(sigma_pri_list), np.array(sigma_sec_list)

# ------------------------------------------
# 1. LOAD CLEAN CSV
# ------------------------------------------

df = pd.read_csv(INPUT_FILE)
print("Loaded clean file:", len(df))

# ------------------------------------------
# 2. RADIUS-GAP SAMPLE
# ------------------------------------------

df = df[
    (df["Rppri"] > R_MIN) & (df["Rppri"] < R_MAX) &
    (df["Rpsec"] > R_MIN) & (df["Rpsec"] < R_MAX)
].copy()

print("Radius-gap sample:", len(df))

# normalize host probabilities safely
host_sum = df["p_pri"] + df["p_sec"]
df = df[host_sum > 0].copy()
df["p_pri"] = df["p_pri"] / (df["p_pri"] + df["p_sec"])
df["p_sec"] = 1.0 - df["p_pri"]

# basic derived quantities
df["delta_radius"] = df["Rpsec"] - df["Rppri"]
df["radius_shift"] = df["Rpsec"] / df["Rppri"]

# ambiguity summary
ambiguous = df[(df["p_pri"] > 0.2) & (df["p_pri"] < 0.8)]
print(f"Ambiguous hosts: {len(ambiguous)} / {len(df)} = {len(ambiguous)/len(df):.3f}")
print("\nRadius shift summary:")
print(df["radius_shift"].describe())

# ------------------------------------------
# 3. BASIC PLOTS
# ------------------------------------------

plt.figure(figsize=(8, 5))
plt.hist(df["p_pri"], bins=20, density=True)
plt.xlabel(r"$p_{\rm pri}$")
plt.ylabel("Density")
plt.title("Host-Probability Distribution")
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR, "p_pri_distribution.png"), dpi=200)
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df["delta_radius"], bins=25, density=True)
plt.xlabel(r"$R_{p,sec} - R_{p,pri}$ [$R_\oplus$]")
plt.ylabel("Density")
plt.title("Radius Inflation")
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR, "delta_radius_distribution.png"), dpi=200)
plt.show()

plt.figure(figsize=(8, 6))
sc = plt.scatter(df["Rppri"], df["Rpsec"], c=df["p_pri"], cmap="viridis", alpha=0.85)
plt.plot([R_MIN, R_MAX], [R_MIN, R_MAX], "--")
plt.axvspan(1.5, 2.0, alpha=0.12)
plt.axhspan(1.5, 2.0, alpha=0.12)
plt.colorbar(sc, label=r"$p_{\rm pri}$")
plt.xlabel(r"$R_{p,pri}$ [$R_\oplus$]")
plt.ylabel(r"$R_{p,sec}$ [$R_\oplus$]")
plt.title("Primary vs Secondary Host Radii")
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR, "Rppri_vs_Rpsec_scatter.png"), dpi=200)
plt.show()

# ------------------------------------------
# 4. POPULATION DISTRIBUTIONS
# ------------------------------------------

R_grid = np.linspace(R_MIN, R_MAX, GRID_N)

P_primary, P_secondary, P_prob, sigma_pri_arr, sigma_sec_arr = build_population(df, R_grid)

stats_primary = valley_metric(R_grid, P_primary)
stats_secondary = valley_metric(R_grid, P_secondary)
stats_prob = valley_metric(R_grid, P_prob)

print("\nValley metrics:")
print("Primary-only   :", stats_primary)
print("Secondary-only :", stats_secondary)
print("Probabilistic  :", stats_prob)

# ------------------------------------------
# 5. MONTE CARLO HOST-MARGINALIZED POPULATION
# ------------------------------------------

bins = np.linspace(R_MIN, R_MAX, 60)
bin_centers = 0.5 * (bins[:-1] + bins[1:])
bin_width = bins[1] - bins[0]

Rppri = df["Rppri"].to_numpy()
Rpsec = df["Rpsec"].to_numpy()
ppri = df["p_pri"].to_numpy()

mc_hists = np.zeros((MC_REALIZATIONS, len(bin_centers)))

for i in range(MC_REALIZATIONS):
    choose_primary = rng.random(len(df)) < ppri
    means = np.where(choose_primary, Rppri, Rpsec)
    sigmas = np.where(choose_primary, sigma_pri_arr, sigma_sec_arr)

    draws = rng.normal(means, sigmas)
    draws = draws[(draws >= R_MIN) & (draws <= R_MAX)]

    counts, _ = np.histogram(draws, bins=bins)
    if counts.sum() > 0:
        mc_hists[i] = counts / (counts.sum() * bin_width)

mc_median = np.median(mc_hists, axis=0)
mc_lo = np.percentile(mc_hists, 16, axis=0)
mc_hi = np.percentile(mc_hists, 84, axis=0)

mc_median = normalize_pdf(mc_median, bin_centers)
mc_lo = normalize_pdf(mc_lo, bin_centers)
mc_hi = normalize_pdf(mc_hi, bin_centers)

mc_stats = valley_metric(bin_centers, mc_median)
print("\nMC median valley metrics:", mc_stats)

# ------------------------------------------
# 6. BOOTSTRAP VALLEY DEPTH
# ------------------------------------------

boot_depths = []

for b in range(BOOTSTRAP_REALIZATIONS):
    boot_idx = rng.integers(0, len(df), len(df))
    boot_df = df.iloc[boot_idx].reset_index(drop=True)

    # rebuild population for bootstrap sample
    Rb = np.linspace(R_MIN, R_MAX, GRID_N)
    _, _, Pboot, _, _ = build_population(boot_df, Rb)
    boot_metric = valley_metric(Rb, Pboot)
    boot_depths.append(boot_metric["depth_ratio"])

boot_depths = np.array(boot_depths)
boot_depths = boot_depths[np.isfinite(boot_depths)]

boot_median = np.median(boot_depths)
boot_lo = np.percentile(boot_depths, 16)
boot_hi = np.percentile(boot_depths, 84)

print("\nBootstrap valley depth ratio:")
print(f"median = {boot_median:.3f}")
print(f"16-84% = [{boot_lo:.3f}, {boot_hi:.3f}]")

# ------------------------------------------
# 7. SAVE RESULTS
# ------------------------------------------

summary = pd.DataFrame([
    {"model": "primary_only", **stats_primary},
    {"model": "secondary_only", **stats_secondary},
    {"model": "probabilistic", **stats_prob},
    {"model": "mc_median", **mc_stats},
])

summary.to_csv(os.path.join(OUTDIR, "valley_summary.csv"), index=False)

mc_df = pd.DataFrame({
    "R_bin_center": bin_centers,
    "mc_median": mc_median,
    "mc_lo": mc_lo,
    "mc_hi": mc_hi,
})
mc_df.to_csv(os.path.join(OUTDIR, "mc_population_summary.csv"), index=False)

pd.DataFrame({"boot_depth_ratio": boot_depths}).to_csv(
    os.path.join(OUTDIR, "bootstrap_valley_depths.csv"),
    index=False
)

with open(os.path.join(OUTDIR, "bootstrap_summary.json"), "w") as f:
    json.dump(
        {
            "boot_median": float(boot_median),
            "boot_lo": float(boot_lo),
            "boot_hi": float(boot_hi),
            "n_boot": int(len(boot_depths)),
        },
        f,
        indent=2,
    )

# ------------------------------------------
# 8. FINAL COMPARISON PLOT
# ------------------------------------------

plt.figure(figsize=(10, 6))
plt.plot(R_grid, P_primary, label="Primary-only", lw=2)
plt.plot(R_grid, P_secondary, label="Secondary-only", lw=2, alpha=0.8)
plt.plot(R_grid, P_prob, label="Probabilistic", lw=2.5)
plt.fill_between(bin_centers, mc_lo, mc_hi, alpha=0.2, label="MC 16–84% band")
plt.plot(bin_centers, mc_median, "--", lw=2, label="MC median")
plt.axvspan(1.5, 2.0, alpha=0.12, label="Gap window")
plt.xlabel(r"Planet Radius [$R_\oplus$]")
plt.ylabel("Normalized Population Density")
plt.title("Binary-Aware Radius Distribution")
plt.xlim(R_MIN, R_MAX)
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTDIR, "binary_aware_radius_distribution.png"), dpi=220)
plt.show()

# ------------------------------------------
# 9. GLOBAL SUMMARY
# ------------------------------------------

print("\n==============================")
print("GLOBAL SUMMARY")
print("==============================")
print("Clean sample size:", len(df))
print("Ambiguous hosts:", len(ambiguous))
print("Ambiguous fraction:", len(ambiguous) / len(df))
print("Mean p_pri:", df["p_pri"].mean())
print("Median p_pri:", df["p_pri"].median())
print("Mean radius shift:", df["radius_shift"].mean())
print("Median radius shift:", df["radius_shift"].median())
print("Primary-only valley depth ratio:", stats_primary["depth_ratio"])
print("Secondary-only valley depth ratio:", stats_secondary["depth_ratio"])
print("Probabilistic valley depth ratio:", stats_prob["depth_ratio"])
print("MC median valley depth ratio:", mc_stats["depth_ratio"])
print(f"Bootstrap median valley depth ratio: {boot_median:.3f}")
print(f"Bootstrap 16-84% range: [{boot_lo:.3f}, {boot_hi:.3f}]")
print(f"\nSaved outputs to: {OUTDIR}")
```
